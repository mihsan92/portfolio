## Data Transformation

Removing unnecassary columns because Indicator Name and Indicator Code columns have only one unique value that won't be needed in data analysis and Unnamed: 69 column has only NA values.

Also, renaming other column names to remove spaces between words

```python
# remove unnecessary columns
world_data = world_data.drop(['Indicator Name','Indicator Code','Unnamed: 69'], axis=1)
#rename column names to get rid of spaces
world_data = world_data.rename(columns={'Country Name':'country_name','Country Code':'country_code'})

world_data.head()
```

Unpivoting data in order to get year values in rows instead of separate columns. 

Creating a country list to be used later in data visualization. 

```python
#unpivot data into by country by year
world_data2 = pd.melt(world_data, id_vars=["country_name", "country_code"], var_name="year", value_name="population")
#change data type of year column to integer
world_data2['year'] = world_data2['year'].astype(int)
```

```python
#unpivot data into by country by year
world_data2 = pd.melt(countries_df, id_vars=["country_name", "country_code"], var_name="year", value_name="population")
#change data type of year column to integer
world_data2['year'] = world_data2['year'].astype(int)
#sort data by year
world_data2 = world_data2.sort_values(by=['year','country_name'],ascending=True)
# divide population by 1000 to display as millions to keep numbers shorter
world_data2['population_000'] = round(world_data2['population'] / 1000,2)

#create a list of 5 random country names
country_list = random.sample(list(world_data2['country_name'].unique()),5)

world_data2.head()
```

```python
# Running an EDA code on the new world_data2 dataframe to get overview of it

def eda_on_df(df):
    print("-"*20+"\nSample Data\n"+"-"*20)
    display(df.sample(5))
    
    print()
    
    print("-"*20+"\nColumns in the dataframe\n"+"-"*20)
    print(df.columns)
    
    print("-"*20+"\nRow and Column Count\n"+"-"*20)
    print(f"Row Count:    {df.shape[0]}\nColumn Count: {df.shape[1]}\n")

    print("-"*20+"\nInfo on the DataFrame\n"+"-"*20)
    print(df.info())

    print("-"*20+"\nNumber of NA values in columns\n"+"-"*20)
    missing_values = df.isna().sum()
    total_values = df.isna().count()
    missing_count = pd.concat([missing_values,total_values],axis=1,keys=['NA','Total'])
    missing_count['perc'] = round((missing_values / total_values) * 100,2)
    display(missing_count.sort_values(by=['perc'],ascending=False))

    print("-"*20+"\nDescriptive Statistics\n"+"-"*20)
    print(df.describe())

    print("-"*20+"\nDescriptive Statistics on the categorical data\n"+"-"*20)
    print(df.describe(include='object'))
    
eda_on_df(world_data2)
```

Selecting 5 random countries to create a line chart visual to display yearly population change by countries. 

It calculates year-over-year population change and computes percentage growth relative to the previous year.

```python
for country in country_list:
    df_filtered = world_data2[world_data2['country_name'] == country]
    country_name = df_filtered['country_name'].unique()
    
    world_data_year = df_filtered.groupby('year')[['population']].agg(population=pd.NamedAgg(column="population", aggfunc="sum")).reset_index()

    world_data_year['population_pre'] = world_data_year['population'].shift(1)
    world_data_year['diff'] = world_data_year['population'] - world_data_year['population_pre']
    world_data_year['diff_perc'] = (world_data_year['population'] - world_data_year['population_pre']) / world_data_year['population']
    
    fig, ax1 = plt.subplots(figsize=(16,4))
    
    df = world_data_year
    
    # Plot line with markers (ticks on each data point)
    ax1.plot(
        df['year'],
        df['diff_perc'],
        marker='.',
        linestyle='-',
        label='Population Change (%)'
    )
    
    # Horizontal line for average
    avg_value = df['diff_perc'].mean()
    ax1.axhline(
        avg_value,
        linestyle='--',
        linewidth=2,
        label=f'Average ({avg_value:.2f})'
    )

    ax1.fill_between(
        df['year'],
        df['diff_perc'],
        avg_value,
        alpha=0.1
    )
    
    # X-axis ticks
    ax1.set_xticks(range(1960, 2030, 5))
    
    # Labels & title
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Population Change (%)')
    ax1.set_title('World Population Change Over Time')
    
    # Grid (background)
    ax1.grid(
        True,
        which='both',
        linestyle='--',
        linewidth=0.5,
        alpha=0.7
    )

    # Country Name Text
    ax1.text(
        0.81, 0.95,                  
        f"{country}",
        transform=ax1.transAxes,      
        fontsize=11,
        verticalalignment='top',
        horizontalalignment='right',
        color="black",
        bbox=dict(
            boxstyle="round,pad=0.3",
            facecolor="white",
            edgecolor="black",
            alpha=0.8
        )
    )

    
    # Legend
    ax1.legend()
    plt.show()
```
